### generar un dataset de pruebas

```python
import pandas as pd
import numpy as np

mean = (0,-4)
C = np.array([[0.3, 0.1], [0.1, 1.5]])
datos1 = np.random.multivariate_normal(mean, C, 500)

#generador de luna
n_samples_out=500
outer_circ_x = np.cos(np.linspace(0, np.pi, n_samples_out))*3
outer_circ_y = np.sin(np.linspace(0, np.pi, n_samples_out))*3

datos2 = np.vstack((outer_circ_x,outer_circ_y)).T
from sklearn.utils import check_random_state
generator = check_random_state(10)
datos2 += generator.normal(scale=0.3, size=datos2.shape)
```

```python
#crea el dataset
X = np.concatenate((datos1, datos2), axis=0)
#construir etiqueta para cada conjunto de datos
n = 20 #ruido/noise
y1 = np.zeros(datos1.shape[0]+n) 
y2 = np.ones(datos2.shape[0]-n)
y =  np.concatenate((y1,y2),axis=0)
```

```python
#verificar frontera
import matplotlib.pyplot as plt
def visualize_border(model,x,y,title=""):
    fig = plt.figure(figsize=(12,6))
    plt.scatter(x[:,0], x[:,1], s=50, c=y, cmap=plt.cm.winter)
    h = .02  # step size in the mesh
    # create a mesh to plot in
    x_min, x_max = x[:, 0].min() - 1, x[:, 0].max() + 1
    y_min, y_max = x[:, 1].min() - 1, x[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, m_max]x[y_min, y_max].
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]) #para logistica

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.contour(xx, yy, Z, cmap=plt.cm.Paired)
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.title(title)
    plt.show()
```

```python
import matplotlib 
def plot_ellipse(splot, mean, cov, color):    
    v, w = np.linalg.eigh(cov)
    u = w[0] / np.linalg.norm(w[0])
    angle = np.arctan(u[1] / u[0])
    angle = 180 * angle / np.pi  
    ell = matplotlib.patches.Ellipse(mean, 3 * v[0] ** 0.5, 3 * v[1] ** 0.5,180 + angle, 
                                     facecolor=color)
    ell.set_clip_box(splot.bbox)
    ell.set_alpha(0.4)
    splot.add_artist(ell)
    
f, axarr = plt.subplots(1, 2,figsize=(12,5))
colors = [plt.cm.winter(float(i)) for i in [0,1]]

axarr[0].scatter(X[y==0][:,0], X[y==0][:,1], s=10, c=colors[0], label="Conjunto 1")
axarr[0].scatter(X[y==1][:,0], X[y==1][:,1], s=10, c=colors[1], label="Conjunto 2")

plot_ellipse(axarr[0], model_lda.means_[0], model_lda.covariance_, colors[0])
plot_ellipse(axarr[0], model_lda.means_[1], model_lda.covariance_, colors[1])
axarr[0].set_title('LDA')


axarr[1].scatter(X[y==0][:,0], X[y==0][:,1], s=10, c=colors[0], label="Conjunto 1")
axarr[1].scatter(X[y==1][:,0], X[y==1][:,1], s=10, c=colors[1], label="Conjunto 2")

plot_ellipse(axarr[1], model_qda.means_[0], model_qda.covariance_[0], colors[0])
plot_ellipse(axarr[1], model_qda.means_[1], model_qda.covariance_[1], colors[1])
axarr[1].set_title('QDA')


plt.show()
```

### mediante el despeje
\begin{align}
e^{\beta_0+\beta_1 x_1 + \beta_2 x_2} &=1 \\
\beta_0+\beta_1 x_1 + \beta_2 x_2 &= 0 \\
x_2 &= - \frac{\beta_0}{\beta_2} - \frac{\beta_1}{\beta_2} x_1 
\end{align}

```python
model = LR(C=0.0001)
model.fit(X,y)
```

```python
coef = model.coef_[0]
interc = model.intercept_[0]
def f(x):
    return -interc/coef[1]-x*coef[0]/coef[1] 

t = np.arange(-5., 5., 0.2)
plt.scatter(X[:,0], X[:,1], s=30, c=y, cmap=plt.cm.winter)
plt.plot(t,f(t))
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```
